```python
exactness= 0.0000001

def f(x):
    return np.sin(x)-0.3+0.5*x

def bisection(xl,xu,N):
    val= (xl+xu)/2
    if N==0:
        return val
    elif f(val) <0: 
        return bisection(val, xu, N-1)
    else: 
        return bisection(xl, val, N-1)

def steigung(x):
    return (f(x)-f(x+0.0000001))/(-0.0000001)

def newtonSolve(x):
    erg= x-f(x)/steigung(x)
    if np.abs(f(erg))<exactness:
        return erg
    else:
        return newtonSolve(erg)
    return

I= [-10,10]
start = bisection(I[0], I[1], 8)
newtonSolve(start)
```

```python
def f(x):
    return np.sin(np.power(x,2))
I=[-3,3]
N=[20, 50, 200]

def vorwDiffQuot(func, x, i):
    if i>=x.size-1 or i<0:
        return 0.0
    return (func(x[i+1])-func(x[i]))/(x[i+1]-x[i])

for k in range(0,3):
    X= np.linspace(I[0], I[1], N[k])
    norm= f(X)
    abl= np.empty(N[k])#, dtype=np.float64)
    for i in range(0,N[k]-1):
        abl[i]= vorwDiffQuot(f, X, i)
    abl[N[k]-1]= -abl[0]
    plt.figure(k)
    plt.plot(X, norm, label="f(x)", c="black")
    plt.plot(X, abl, label="f'(x)", c="red")
    plt.grid()
    plt.legend()

def fPrime(x):
    return np.cos(np.power(x,2))*2*x

def zentDiffQuot(func, x, i):
    if i>=x.size-1 or i<1:
        return 0
    return (func(x[i+1])-func(x[i-1]))/((x[i+1]-x[i-1]))


Nb= 100
X= np.linspace(I[0], I[1], Nb)
FP= fPrime(X)
FV= np.zeros(Nb)
FZ= np.zeros(Nb)
for i in range(0,Nb):
    FV[i]= vorwDiffQuot(f, X, i)
    FZ[i]= zentDiffQuot(f, X, i)
DV= np.sum(np.abs(FV-FP))/Nb
DZ= np.sum(np.abs(FZ-FP))/Nb
print("DV= "+str(DV))
print("DZ= "+str(DZ))
plt.figure(4)
plt.plot(X, FP, label="FP")
plt.plot(X, FV, label="FV")
plt.plot(X, FZ, label="FZ")
plt.grid()
plt.legend()
```

$ m \cdot \frac {d^2}{dt^2} \xi_1 = k \cdot (- \xi_1 + \xi_2) $

$ m \cdot \frac {d^2}{dt^2} \xi_N = k \cdot (- \xi_N + \xi_{N-1}) $

$ m \cdot \frac {d^2}{dt^2} \xi_n = k \cdot (-2 \xi_n + \xi_{n-1} + \xi_{n+1}) $

$ m \cdot \frac {d^2}{dt^2} \vec{\xi}(t) = k M \vec{\xi}(t) $

$ \Rightarrow M = \begin{pmatrix}
-1 & 1& 0 & 0 & \cdots & 0 & 0 & 0 \\
1 & -2 & 1 & 0 & \cdots & 0 & 0 & 0 \\
0 & 1 & -2 & 1 & \cdots & 0 & 0 & 0 \\
\vdots & \vdots & \vdots & \vdots & \ddots & \vdots & \vdots & \vdots \\
0 & 0 & 0 & 0 & \cdots & 1 & -2 & 1 \\
0 & 0 & 0 & 0 & \cdots & 0 & 1 & -1
\end{pmatrix}$

$ \vec{\xi}(t) = \sum_{n=1}^N \vec{a_n} (\alpha_n cos(\omega_n t) + \beta_n sin(\omega_n t)) $

$ (\vec{\xi}(t))_n = \xi_n(t) $

$ \xi_1(0)=1 \quad \Rightarrow \quad \sum_{n=1}^N (\vec{a_n})_1 (\alpha_n cos(0) + \beta_n sin(0))=1  \quad \Rightarrow \quad \sum_{n=1}^N (\vec{a_n})_1 \alpha_n=1 $

$ \xi_i(0)=0 \quad i=2,...,N \quad \Rightarrow \quad \sum_{n=1}^N (\vec{a_n})_i (\alpha_n cos(0) + \beta_n sin(0))=0  \quad \Rightarrow \quad \sum_{n=1}^N (\vec{a_n})_i \alpha_n=0 $

$ \dot \xi_i(0)=0 \quad i=1,...,N \quad \Rightarrow \quad \sum_{n=1}^N (\vec{a_n})_i (-\alpha_i \omega_i sin(0) + \beta_i \omega_i cos(0))=0 \quad \Rightarrow \quad \sum_{n=1}^N (\vec{a_n})_i \beta_n=0 $

$ \Rightarrow \quad \xi_m(t) = \sum_{n=1}^N (\vec{a_n})_m (\alpha_n cos(\omega_n t) + \beta_n sin(\omega_n t)) = \sum_{n=1}^N (\vec{a_n})_m \alpha_n cos(\omega_n t) + \sum_{n=1}^N (\vec{a_n})_m \beta_n sin(\omega_n t) $

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

N= 50
dt= 0.02

M= np.zeros((N, N))
for i in range(1,N):
    M[i,i]= -2
    M[i-1,i]= 1
    M[i,i-1]= 1
M[0,0]= -1
M[N-1, N-1]= -1

om, EV= np.linalg.eig(M)

om= np.sqrt(np.abs(om))
#om= sorted(om)[::-1]
plt.figure(1, figsize=[16,9])
T= np.linspace(0,10,501) 
for m in range(0,5):
    L= np.zeros(501)
    for i in range(0,501):
        for n in range(0,N):
            t= i*dt
            L[i]+= EV[m,n]*EV[1,n]*np.cos(om[n]*t)
    plt.plot(T, L)
plt.grid()
```

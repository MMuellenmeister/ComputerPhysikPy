```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

N = 3
A0 = np.random.rand(N,N)
A0=np.floor(A0*100)
b0 = np.random.rand(N)
b0=np.floor(b0*100)
def gausssolve(A0,b0):
    N=b0.shape[0]
    x = np.empty(N)
    print("Ausgangsmatrix A0:\n", A0)
    print("b0 =", b0)
    
    A=   np.copy(A0)
    Ap1= np.copy(A0)
    b=   np.copy(b0)
    bp1= np.copy(b0)
    
    # Teil 1: Transformation auf obere Dreiecksmatrix
    for n in range(0,N-1):
        for i in range(n+1,N):
            factor = A[i,n]/A[n,n]
            for j in range(0,N):
                Ap1[i,j] = A[i,j] - factor*A[n,j]
            bp1[i] = b[i] - factor*b[n]
        A = np.copy(Ap1)
        b = np.copy(bp1)
    print("obere Dreiecksmatrix A^(N-1):\n"+str(A))
    
    for i in range(N-1,-1,-1):
        s= 0
        for k in range(N-1, i, -1):
            s+=A[i,k]*x[k]
        x[i]=(b[i]-s)/(A[i,i])
    return x
x= gausssolve(A0,b0)
bGes= np.zeros(N)
for i in range(0,N):
    for k in range(0,N):
        bGes[i]+=A0[i,k]*x[k]
print(bGes)
will=gausssolve(np.array([[0.5, -2, 2], [2, 1, -4], [-1, -1, 2]]), np.array([3/2, 0, 3]))
print(will)
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

def E0(n, theta=0):
    return n+1/2-theta**2

def Hmat(theta, N):
    H=np.zeros((N,N))
    for i in range(0,N):
        H[i,i]= E0(i,theta)
        if i<N-1:
            H[i, i+1]= np.sqrt(i+1)*theta
            H[i+1, i]= np.sqrt(i+1)*theta
    return H
N=10
plt.figure(1, figsize=[4.5,4.5])
H=Hmat(1, N)
plt.matshow(H, fignum=1)

plt.figure(2, figsize=[16,9])
plt.grid()
dim=1+20
Theta= np.linspace(0,2,dim)
SolAna= np.empty((dim,N))
SolNum= np.empty((dim,N))
for t in range(0,dim):
    SolNum[t]= np.linalg.eigvals(Hmat(Theta[t], N))
    for n in range(0,N):
        SolAna[t][n]= E0(n, theta=Theta[t])
for i in range(0,N):
    col= "#%x" % np.abs(np.floor((i+1)*2**(4*6)/(N+1))).astype(int)
    plt.plot(Theta, SolNum[:,i], lw=2, c= col)
    plt.plot(Theta, SolAna[:,i], lw=2, c= col, ls=":")
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

def H(x, n):
    if n==0:
        return 1.0
    if n==1:
        return 2*x
    return 2*x*H(x,n-1)-2*n*H(x,n-2)

def psi(x, n):
    return np.power(np.pi, -(1/4))*1/np.sqrt(np.power(2,n)*np.math.factorial(n))*H(x, n)*np.exp(-1/2*np.power(x, 2))

N= np.linspace(0,4,5, dtype=int)
X= np.linspace(-5,5,201)
Sol= np.zeros((N.shape[0],X.shape[0]))
print(Sol.shape)
plt.figure(2, figsize=[10,10])
for a in range(0,5):
    for b in range(0,201):
        Sol[a,b]= psi(X[b], N[a])+N[a]+0.5
    plt.plot(X, Sol[a], label= "n="+str(N[a]))
plt.plot(X, np.power(X,2)/5, c="black")
plt.grid()
plt.legend()
```

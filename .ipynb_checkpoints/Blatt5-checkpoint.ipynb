Umformen von den Differenzialgeichungen

$m\ddot{r}=mr\dot{\phi}^2-\frac{\partial V}{\partial r}\qquad (1)$

$\frac{d}{dt}(mr^2\dot{\phi})=mr^2\ddot{\phi}+2mr\dot{r}\dot{\phi}=0\qquad (2)$

Umstellen:

$\ddot{r}=r\dot{\phi}^2-\frac{1}{m}\frac{\partial V}{\partial r} \qquad (1)$

$\ddot{\phi}=-\frac{2\dot{r}\dot{\phi}}{r} \qquad (2)$

Setze $\quad \vec{u}(t) \: =\;
    \begin{pmatrix}
          u_1(t) \\
          u_2(t) \\
          u_3(t) \\
          u_4(t) \\
    \end{pmatrix}
    \; =\;
    \begin{pmatrix}
          r(t) \\
          \phi (t) \\
          \dot{r}(t) \\
          \dot{\phi}(t) \\
    \end{pmatrix}
    \quad \Rightarrow \quad
    \dot{\vec{u}}(t) \: = \;
    \begin{pmatrix}
          \dot{r}(t) \\
          \dot{\phi} (t) \\
          \ddot{r}(t) \\
          \ddot{\phi}(t) \\
    \end{pmatrix}
    \; =\;
    \begin{pmatrix}
        u_3(t) \\
        u_4(t) \\
        g_1(\vec{u},t) \\
        g_2(\vec{u},t) \\
    \end{pmatrix}
    $
    
mit

$g_1(\vec{u},t)= u_1u_4^2-\frac{1}{m}\frac{\partial V}{\partial u_1}$

$g_2(\vec{u},t)=-\frac{2u_3u_4}{u_1}$

Daraus folgt also:

$\Rightarrow \quad \frac{d\vec{u}}{dt}=\vec{f}(\vec{u},t) \: = \;
\begin{pmatrix}
    u_3(t) \\
    u_4(t) \\
    g_1(\vec{u},t) \\
    g_2(\vec{u},t) \\
\end{pmatrix}
$

Reduzieren der Differenzialglaichung

$\ddot{\varphi}(t)+\frac{1}{l(t)}\sin(\varphi(t))+2\frac{\dot{l}(t)}{l(t)}\dot{\varphi}(t)=0 \qquad (1)$

$l(t)=1+a\:sin(2t) \qquad (2)$

Umstellen:

$\ddot{\varphi}(t) = -\frac{1}{l(t)}\sin(\varphi(t))-2\frac{\dot{l}(t)}{l(t)}\dot{\varphi}(t)=0 \qquad (1)$

>*Anmerkung: da $l(t)$ schon explizit gegeben ist, kÃ¶nnen die Ableitungen $\dot{l}(t)$ und >$\ddot{l}(t)$ schon einfach analytisch berechnet werden:*
>
>$\Rightarrow \quad \dot{l}(t)= 2a \cos(2t)$
>
>$\Rightarrow \quad \ddot{l}(t)= -4a \sin(2t)$

Setze 

$\qquad \, \vec{u}(t)\: = \;
\begin{pmatrix}
    u_1(t) \\
    u_2(t) \\
    u_3(t) \\
    u_4(t) \\
\end{pmatrix}
\; =\;
\begin{pmatrix}
    \varphi (t) \\
    \dot{\varphi}(t) \\
    l(t) \\
    \dot{l}(t) \\
\end{pmatrix}
$

$\Rightarrow \quad \dot{\vec{u}}(t) \: = \; 
\begin{pmatrix}
    \dot{u_1}(t) \\
    \dot{u_2}(t) \\
    \dot{u_3}(t) \\
    \dot{u_4}(t) \\
\end{pmatrix}
\; =\;
\begin{pmatrix}
    \dot{\varphi} (t) \\
    \ddot{\varphi}(t) \\
    \dot{l}(t) \\
    \ddot{l}(t) \\
\end{pmatrix}
\; =\;
\begin{pmatrix}
    u_2(t) \\
    f(\vec{u},t) \\
    u_4(t) \\
    \ddot{l}(t) \\
\end{pmatrix}
$

mit $\; f( \vec{u},t)= - \frac{1}{u_3} \sin(u_1)-2 \frac{u_4}{u_3}u_2 \qquad (vgl \: 1)$

>*Da l jedoch schon gegeben ist, ist der relevante Teil nur*
$ \:
\begin{pmatrix}
    \dot{\varphi} (t) \\
    \ddot{\varphi}(t) \\
\end{pmatrix}
\; =\;
\begin{pmatrix}
    u_2(t) \\
    f(\vec{u},t) \\
\end{pmatrix}
$

```python
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt

g= 1
psi_0= 0
psiP_0=0.7
dt= 0.01
Trange= [0,10]
dim= np.ceil((Trange[1]-Trange[0])/dt).astype(np.int64)
T= np.linspace(Trange[0],Trange[1], dim)
A= [0.0, 0.1, 0.2, 0.3, 0.4]

def l(a, t):
    return 1+a*np.sin(2*t)
def lP(a, t):
    return 2*a*np.cos(2*t)
def l2p(a, t):
    return -4*a*np.sin(2*t)

def F(uArray, t_index, u_index, dt, aParam):
    t= t_index*dt
    if u_index == 0:
        return uArray[1][t_index]
    elif u_index == 1:
        return -1/l(aParam, t)*(np.sin(uArray[0][t_index])+2*lP(aParam, t)*u[1][t_index])
    return None

def EulerVerf(func, u, start, dt, FuncParams= None):
    """
    Uses Euler-method to solve partial differential eqn func in Array u. 
    @param
        func: Function to be solved. Must have parameters: 
            uArray (same as u)  
            t_index (second index of u) 
            u_index (first index of u) 
            dt (smae as dt) 
            aParam (same as FuncParams)
        u: NumPy Array in which the function is solved. must have 2 dimensions: 
            first one for the partial differential equations u_1, u_2, etc., 
            second one for the steps in time, i.e. u_1(X)
        start: List with starting values of u, i.e. [u_1(0), u_2(0), ...]
        dt: time step between each entry of the PDE's
        FuncParams: for additional arguments required for func
    @returns None, edits u in-place
    """
    for i in range(0,len(start)):
        u[i][0]= start[i]
    for i in range(1,u.shape[-1]):
        for k in range(0,u.shape[0]):
            u[k][i]= u[k][i-1]+dt*func(u, i-1, k, dt, FuncParams)
    return

for k in range(0, len(A)):
    a= A[k]
    u= np.empty((2, dim))
    EulerVerf(F, u, [psi_0, psiP_0], dt, a)
    plt.plot(T, u[0])
```

```python
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt

t_vals= [0.0, 10.0]
steps= [250, 1000, 4000]
x_start= [1, 0]

def EulerVerf(func, u, start, dt, FuncParams= None):
    """
    Uses Euler-method to solve partial differential eqn func in Array u. 
    @param
        func: Function to be solved. Must have parameters: 
            uArray (same as u)  
            t_index (second index of u) 
            u_index (first index of u) 
            dt (smae as dt) 
            aParam (same as FuncParams)
        u: NumPy Array in which the function is solved. must have 2 dimensions: 
            first one for the partial differential equations u_1, u_2, etc., 
            second one for the steps in time, i.e. u_1(X)
        start: List with starting values of u, i.e. [u_1(0), u_2(0), ...]
        dt: time step between each entry of the PDE's
        FuncParams: for additional arguments required for func
    @returns None, edits u in-place
    """
    for i in range(0,len(start)):
        u[i][0]= start[i]
    for i in range(1,u.shape[-1]):
        for k in range(0,u.shape[0]):
            u[k][i]= u[k][i-1]+dt*func(u, i-1, k, dt, FuncParams)
    return
def f(uArray, t_index, u_index, dt, aParam):
    if u_index ==0:
        return -1*uArray[1][t_index]
    if u_index ==1: 
        return uArray[0][t_index]
    return None

for N in steps:
    dt= (t_vals[1]-t_vals[0])/N
    u= np.empty((2, N))
    EulerVerf(f, u, x_start, dt)
    #plt.figure(N)
    T= np.linspace(t_vals[0], t_vals[1], N)
    plt.plot(T, u[0], label= "N="+str(N))
plt.plot([0,10],[0,0], c= "black")
plt.plot([0,10],[-1,-1], c= "black")
plt.plot([0,10],[1,1], c= "black")
plt.legend()
plt.grid()
plt.xlim((0,10))
```

```python
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt

t_vals= [0.0, 10.0*np.pi] #start and stop values for the time
N=500 #number of steps for every method
x_start= [1, 0] #starting value of the DE's, first one i x(t), then x'(t) and so on

def EulerVerf(func, u, start, dt, FuncParams= None):
    """
    Uses Euler-method to solve partial differential eqn func in Array u. 
    @param
        func: Function to be solved. Must have parameters: 
            uArray (same as u)  
            t_index (second index of u) 
            u_index (first index of u) 
            dt (smae as dt) 
            aParam (same as FuncParams)
        u: NumPy Array in which the function is solved. must have 2 dimensions: 
            first one for the partial differential equations u_1, u_2, etc., 
            second one for the steps in time, i.e. u_1(X)
        start: List with starting values of u, i.e. [u_1(0), u_2(0), ...]
        dt: time step between each entry of the DE's
        FuncParams: for additional arguments required for func
    @return None, edits u in-place
    """
    for i in range(0,len(start)):
        u[i][0]= start[i] #set up starting values
    for i in range(1,u.shape[-1]): #go through every (time) step for func
        for k in range(0,u.shape[0]): #go through every eqn of the system
            u[k][i]= u[k][i-1]+dt*func(u, i-1, k, dt, FuncParams) #solve eqn via Euler
    return None

def RK2solve(func, u, start, dt, FuncParams= None):
    """
    Uses second order Runge-Kutta to solve the DE func in Array u. 
    @param
        func: Function to be solved. Must have parameters: 
            uArray (same as u)  
            t_index (second index of u) 
            u_index (first index of u) 
            dt (smae as dt) 
            aParam (same as FuncParams)
        u: NumPy Array in which the function is solved. must have 2 dimensions: 
            first one for the partial differential equations u_1, u_2, etc., 
            second one for the steps in time, i.e. u_1(X)
        start: List with starting values of u, i.e. [u_1(0), u_2(0), ...]
        dt: time step between each entry of the DE's
        FuncParams: for additional arguments required for func
    @returns None, edits u in-place
    """
    for i in range(0,len(start)):
        u[i][0]= start[i] #set up starting values
    for i in range(1,u.shape[-1]): #go through every (time) step for func
        for k in range(0,u.shape[0]): #go through every eqn of the system
            fi = f(u, i-1, k, dt, FuncParams) #calculate first increment
            a = f(u + dt*fi, i-1, k, dt, FuncParams) #calculate second increment
            u[k][i]= u[k][i-1] + 0.5*dt*fi + 0.5*dt*a #calculate next value of solution
    return None

def RK4solve(func, u, start, dt, FuncParams= None):
    """
    Uses fourth order Runge-Kutta to solve the DE func in Array u. 
    @param
        func: Function to be solved. Must have parameters: 
            uArray (same as u)  
            t_index (second index of u) 
            u_index (first index of u) 
            dt (smae as dt) 
            aParam (same as FuncParams)
        u: NumPy Array in which the function is solved. must have 2 dimensions: 
            first one for the partial differential equations u_1, u_2, etc., 
            second one for the steps in time, i.e. u_1(X)
        start: List with starting values of u, i.e. [u_1(0), u_2(0), ...]
        dt: time step between each entry of the DE's
        FuncParams: for additional arguments required for func
    @returns None, edits u in-place
    """
    for i in range(0,len(start)):
        u[i][0]= start[i] #set up starting values
    for i in range(1,u.shape[-1]): #go through every (time) step for func
        for k in range(0,u.shape[0]): #go through every eqn of the system
            c1=dt*func(u, i-1, k, dt, FuncParams) #calculate first increment
            c2=dt*func(u+0.5*c1, i-1, k, dt, FuncParams) #calculate second increment
            c3=dt*func(u+0.5*c2, i-1, k, dt, FuncParams) #calculate third increment
            c4=dt*func(u+c3, i-1, k, dt, FuncParams) #calculate fourth increment
            u[k][i]= u[k][i-1]+(c1+2.0*c2+2.0*c3+c4)/6.0 #calculate next value of solution
    
    return None

def f(uArray, t_index, u_index, dt, aParam):
    #the parameters for this function are a bit strange because of the other exercises, 
    # for which i want to reuse this code
    if u_index ==0:
        return -1*uArray[1][t_index]
    if u_index ==1: 
        return uArray[0][t_index]
    return None

T= np.linspace(t_vals[0], t_vals[1], N) #time-axisa
u= np.empty((2,N)) #solution-array for the Euler method
cos= np.cos(T) #exact solution
dt= (t_vals[1]-t_vals[0])/N #time step(not strictly needed here but required for this implementation)

EulerVerf(f, u, x_start, dt)
v= np.empty((2,N))#solution-array for the RK4 method
RK4solve(f, v, x_start, dt)
w= np.empty((2, N))#solution-array for the RK2 method
RK2solve(f, w, x_start, dt)
#plotting
plt.plot(T, cos, label="Exact")
plt.plot(T, u[0], label="Euler")
plt.plot(T, v[0], label="RK4")
plt.plot(T, w[0], label="RK2")
plt.grid()
plt.legend()
```

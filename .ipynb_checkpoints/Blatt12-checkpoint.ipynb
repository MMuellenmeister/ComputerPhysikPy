```python
#%%timeit 19.3 s  72.2 ms per loop (mean  std. dev. of 7 runs, 1 loop each)
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

k_B=1

def fullSpinConfig(N):
    sp= np.ones((2**N,N ))
    for i in range(0, 2**N):
        for j in range(0,N):
            if (i >> j) & 1==1:
                sp[i,j]*=-1
    return sp/2

def M(T, N, J, h):
    spins= fullSpinConfig(N)
    beta= 1/(k_B*T)
    M=0.0
    Z=0.0
    for i in range(0,2**N):
        conf= spins[i]
        confSum= np.sum(conf)
        hamilton= 0.0
        for k in range(0, N-1):
            hamilton+= conf[k]*conf[k+1]
        hamilton*=(-J)
        hamilton-=h*confSum
        
        omega= np.exp(-beta*hamilton)
        Z+= omega
        M+=confSum*omega
    return M/Z

J= 1
N= 12
T= np.linspace(0.1,5,100)
hVals= [0.1, 0.2, 0.3, 1.0]
plt.figure(0,figsize=[16,9])
for h in hVals:
    #print("h=",h)
    erg= np.empty(T.shape[0])
    for i in range(0, T.shape[0]):
        #print(i+1,"%")
        erg[i]= M(T[i], N, J, h)
    plt.plot(T, erg, label="h= "+str(h))
plt.xlabel("Temperatur T")
plt.ylabel("Magnetisierung M")
plt.grid()
plt.legend()
#plt.show()
"""#"""
```
